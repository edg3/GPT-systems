```python
import fitz  # PyMuPDF

def extract_paragraphs_from_pdf(pdf_path):
    doc = fitz.open(pdf_path)
    paragraphs = []

    for page in doc:
        # Extract text from the page
        page_text = page.get_text("text")
        # Split the text into paragraphs (you can adjust the delimiter as needed)
        page_paragraphs = page_text.split("\n\n")  # Assuming paragraphs are separated by two newlines
        paragraphs.extend(page_paragraphs)

    return paragraphs

pdf_file_path = '00.RS Digital Multimeter RS-12.A700000011026605.pdf'
extracted_paragraphs = extract_paragraphs_from_pdf(pdf_file_path)
print(extracted_paragraphs)
```
